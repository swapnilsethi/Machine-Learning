#### Question - Design Neural Network with n =5, c = 3 and 1 hidden layer with 4 perceptron. Use Sigmoid Activation Function. Assume suitable weights and bias.

#### Solution - 

Load required libraries

```python
import numpy as np
```

Inputs

```python
X= np.array([[11,12,13],[21,22,23],[31,32,33],[41,42,43],[51,52,53]])
```

```python
print("Inputs are: ")
X
```

#### There are 2 different ways commanly followed in industry to define weights. We will look at both ways.

##### 1. Weights defined in a c by z format, where c represents # of columns of input vectors  and z represents number of perceptron in hidden layer

```python
W1 = np.array([[0.01,0.02,0.03,0.04],[0.02,0.01,0.04,0.02],[0.03,0.01,0.04,0.02]])
```

```python
print("Weights to calculate hidden layer 1 input: ")
W1
```

##### 2. Weights defined in a z by c format, where z represents # of perceptron in hidden layer and c represents # of columns in input vectors.

```python
W1 = np.array([[0.01,0.02,0.03],[0.02,0.01,0.01],[0.03,0.04,0.04],[0.04,0.02,0.02]])
```

```python
W1
```

```python
W1 = W1.transpose()  ## Here we need to take a transpose of weights matrix to format it in proper shape
```

```python
W1
```

```python
b1 = [0.5,0.8,0.3,0.1]
```

```python
b1
```

```python
Z1 = np.dot(X,W1)+b1
```

```python
Z1
```

```python
H1 = 1/(1+np.exp(Z1))
```

```python
print("O/P of hidden layer is:")
H1
```

##### Weights to calculate Y^ from hidden layer outputs

```python
W2 = [[0.01],[0.02],[0.04],[0.03]]
```

```python
W2
```

Bias for o/p layer

```python
b2 = [[0.2]]
```

```python
Z2 = np.dot(H1,W2)+b2  
```

```python
Z2
```

```python
Y_hat = 1/(1+np.exp(Z2))  ## final layer o/p using activation function
```

```python
Y_hat
```
